```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("Iris.csv")
```

```python
#1. List down the features and their types (e.g., numeric, nominal) available in the dataset.  
print("Feature Types:\n")
print(df.dtypes)

# Get unique values to identify nominal vs. numerical
print("\nBasic Summary:\n")
print(df.describe(include='all'))
```

```python
#2. Create a histogram for each feature in the dataset to illustrate the feature distributions.  
```

```python
###############3  HISTOGRAM OF SEPAL LENGTH

plt.figure(figsize = (10,7))
x = df["SepalLengthCm"]
plt.hist(x, bins = 30, color = 'purple')
plt.title("Sepal Length in Cm")
plt.xlabel("Sepal Length")
plt.xlabel("Frequency")
```

```python
###############3  HISTOGRAM OF SEPAL Width

plt.figure(figsize = (10,7))
x = df["SepalWidthCm"]
plt.hist(x, bins = 30, color = 'purple')
plt.title("Sepal Width in Cm")
plt.xlabel("Sepal Width")
plt.xlabel("Frequency")
```

```python
###############3  HISTOGRAM OF PETAL LENGTH

plt.figure(figsize = (10,7))
x = df["PetalLengthCm"]
plt.hist(x, bins = 30, color = 'purple')
plt.title("Petal Length in Cm")
plt.xlabel("Petal Length")
plt.xlabel("Frequency")
```

```python
###############3  HISTOGRAM OF PETAL Width

plt.figure(figsize = (10,7))
x = df["PetalWidthCm"]
plt.hist(x, bins = 30, color = 'purple')
plt.title("Petal Width in Cm")
plt.xlabel("Petal Width")
plt.xlabel("Frequency")
```

```python
#3. Create a boxplot for each feature in the dataset.  
```

```python
# Box plot fot sepal length

plt.figure(figsize = (10,7))
df.boxplot(column = 'SepalLengthCm')           
```

```python
# Box plot fot sepal width

plt.figure(figsize = (10,7))
df.boxplot(column = 'SepalWidthCm')   
```

```python
# Box plot fot petal length

plt.figure(figsize = (10,7))
df.boxplot(column = 'PetalLengthCm')   
```

```python
# Box plot fot petal width

plt.figure(figsize = (10,7))
df.boxplot(column = 'PetalWidthCm')  
```

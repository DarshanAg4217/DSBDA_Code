```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#############   BASIC DATA OPERATIONS
df = pd.read_csv("train.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.describe(include = "object")
##  generate descriptive statistics specifically for categorical (non-numeric) columns in a DataFrame.
```

```python
###############  checking null values
df.isnull().sum()
```

```python
##################  REPLACING NULL VALUES of age by its mean
```

```python
df['Age'] = df['Age'].fillna(np.mean(df['Age']))
df['Age']
```

```python
### CHECK THE NULL VALUES OF AGE AFTER 
df['Age'].isnull().sum()
```

```python
##################  REPLACING NULL VALUES of age by its mode
df['Cabin'] = df['Cabin'].fillna(df['Cabin'].mode()[0])
df['Cabin']
```

```python
print("Null values after")
df['Cabin'].isnull().sum()
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Embarked']
```

```python
print("Null values after")
df['Embarked'].isnull().sum()
```

```python
##### FINAL CHECK NULL VALUES
df.isnull().sum()
```

```python
df_numeric = df.select_dtypes(include=[np.number])
#####      because corr() selects only numeric data
# Compute correlation
tc = df_numeric.corr()

# Plot heatmap
sns.heatmap(tc, cmap='YlGnBu', annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Checking price of tickets using histogram
sns.histplot(data = df, x = 'Fare', bins = 20, hue = 'Pclass')
plt.show()
```
